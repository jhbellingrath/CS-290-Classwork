```python
import pandas as pd
import math
import scipy.stats as stats
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/jhbellingrath/CS-290-Classwork/refs/heads/main/data/breast-cancer%20(4).csv")
df.head()
```

```python
attributes=df.columns[ df.columns !="diagnosis"]
attributes=attributes.tolist()
attributes
```

```python
vals=df['diagnosis'].unique()
vals=vals.tolist()
gaussian_matrix=[]
for val in vals:
  restricted_attributes=df[df['diagnosis']==val]
  temp_matrix=[]
  for attribute in attributes:
    mean=restricted_attributes[attribute].mean()
    std=restricted_attributes[attribute].std()
    temp_matrix.append([mean,std])
  gaussian_matrix.append(temp_matrix)
```

```python
def probability(target_value,attribute,attribute_value):
  i_target=vals.index(target_value)
  i_attribute=attributes.index(attribute)
  z_score=(gaussian_matrix[i_target][i_attribute][0]-attribute_value)/(gaussian_matrix[i_target][i_attribute][1])
  return stats.norm.pdf(z_score)
```

```python
def Naive_Beyes(df):
  probabilities=[]
  best=[]
  columns=df.columns
  for i in range(len(df)):
    for val in vals:
      prob=1
      for column in columns:
        prob=prob*probability(val,column,df[column][i])
      probabilities.append([prob,val])
    probabilities.sort
    best.append(probabilities[0][1])
  return best
```

```python
df=pd.DataFrame({"radius_mean": [-3,-1.1], "perimeter_mean": [-9,-3] })
print(Naive_Beyes(df))
```
